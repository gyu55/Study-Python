```python
chat_template = ChatPromptTemplate.from_messages([
    ("system", "너는 지리학 전문가야! 그리고 오직 {language}언어로 답변해줘"),
    ("human", "{country_a}에서 {country_b}까지의 거리를 알려줘."),
])
```

```python
chat_template
```

```python
chat_prompt = chat_template.format_messages(country_a="한국", country_b="미국", language="한국어")

chat_prompt
```

```python
result = chat.invoke(chat_prompt)
```

```python
result
```

## OutParser

```python
# LLM -> 파이썬 List
from langchain_core.output_parsers.base import BaseOutputParser

# OutParser를 제작하기 위해선 BaseOutputParser를 상속받아야 한다.
```

```python
class CommaOutputParser(BaseOutputParser):
    # parse(): 반드시 재정의
    def parse(self, text):
        return list(map(str.strip, text.split(",")))
```

```python
comma_parser = CommaOutputParser()
```

```python
comma_parser.parse("안녕,             ,나는,                홍 길   동, 이야")
```

```python
template = ChatPromptTemplate.from_messages([
    ("system", """
        리스트를 생성하는 기계입니다.
        요청한 모든 리스트에 개수는 최대 {max_length}개까지만 목록으로 표시하세요.
        그 이상 초과되는 리스트는 답변하지 마세요.
    """),
    ("human", "{question}")
])
```

```python
prompt = template.format_messages(
    max_length=5,
    question="태양계 행성의 종류를 알려줘"
)
```

```python
print(prompt)
```

```python
result = chat.invoke(prompt)
```

```python
output_list = comma_parser.parse(result.content)

print(output_list)
for i in output_list:
    print(i)
```
