# 생성 증강
## RetrievalQA
1) Stuff documents chain
    * 모든 검색 문서를 한 번에 프롬프트에 “다 넣어서” 한 번에 답변 생성합니다
    문서가 적을 때 빠르고 단순하지만, 길면 토큰 초과 위험 있습니다.
    - 특징: 다 몰아서 한 번에

2) Refine documents chain
    - 첫 문서로 1차 답변 생성 → 다음 문서로 점진적으로 답변 보완합니다.
    문서가 많아도 안정적이지만 속도가 느립니다.
    - 특징: 점진적 보완

3) Map reduce documents chain
    - 각 문서마다 개별 답변 생성(Map)합니다. 그것들을 다시 종합(Reduce)합니다.
    따라서 대용량 문서 처리에 적합니다.
    - 특징: 나눠서 만들고 합침

4) Map re-rank documents chain
    - 각 문서별로 답변 + 점수를 생성하고 가장 점수 높은 답변 선택합니다
    "가장 관련성 높은 하나"를 뽑을 때 유리합니다.
    - 특징: 점수 매겨서 최고만 선택

### 4. 프롬프트 작성

```python
# 최초 프롬프트
initial_prompt = ChatPromptTemplate.from_template("""
    제공된 context에서만 질문에 답하세요.
    context:
        {context}

    Answer the question:
        {question}
""")

# refine 프롬프트
refine_prompt = ChatPromptTemplate.from_template("""
    당신의 임무는 최종 요약본을 작성하는 것입니다.
    지금까지의 요약된 내용이 있습니다.
    exsisting_answer:
        {exsisting_answer}

    이제 요약을 개선하는 데 도움이 될 수 있는 추가 context가 있습니다.
    context:
        {context}
        
    새로운 context를 사용해서 원본을 요약하고 다듬습니다.
    새로운 context에서 유용한 정보가 없다면, 원래 요약을 그대로 반환하세요.
""")
```

```python
def refine_docs(inputs):
    docs, question = inputs["docs"], inputs["question"]

    # 첫 번째 문서로 시작
    chain = initial_prompt | chat
    result = chain.invoke({
        "context": docs[0].page_content,
        "question": question
    }).content

    # 두 번째 문서부터 refine 적용
    refine_chain = refine_prompt | chat
    for doc in docs[1:]:
        result = refine_chain.invoke({
            "exsisting_answer": result,
            "context": doc.page_content,
            "question": question
        }).content

    return result
```

```python
refine_docs({
    "docs": docs[:3],
    "question": "김첨지의 직업은?"
})
```

```python
refine_chain = {
    "docs": retrieval,
    "question": RunnablePassthrough()
} | RunnableLambda(refine_docs)
```

```python
final_result = refine_chain.invoke("김첨지가 먹은 음식은?")
```

# Map re-rank documents chain

```python
import re
```

```python
class MapReRankChain:
    def __init__(self, llm, retriever):
        self.llm = llm
        self.retriever = retriever

        # 1. 문서의 관련성 평가하는 프롬프트
        self.re_rank_prompt = ChatPromptTemplate.from_template("""
            다음 context의 질문과 얼마나 관련있는지
            0-10까지의 척도 내에서 숫자로만 대답하세요.
    
            context: 
                {context}
                
            question: 
                {question}
        """)
    
        self.final_prompt = ChatPromptTemplate.from_template("""
            아래 context에서 가장 관련성이 높은 정보를 바탕으로 question에 답하세요.
    
            context:
                {context}

            question:
                {question}
        """)

        self.re_rank_chain = self.re_rank_prompt | self.llm | StrOutputParser()
        self.final_chain = self.final_prompt | self.llm | StrOutputParser()

    # invoke 메서드 생성
    def invoke(self, question: str):
        docs = self.retriever.invoke(question)

        # re_rank 문서의 점수를 매김
        scored_docs = []
        for doc in docs:
            score_str = self.re_rank_chain.invoke({"context": doc.page_content, "question": question})

            # 숫자만 추출
            score = float(re.search(r"\d+", score_str).group()) if re.search(r"\d+", score_str) else 0
            scored_docs.append((score, doc))

        # 정렬 상위 k개
        scored_docs.sort(key=lambda tu: tu[0], reverse=True)
        top_docs = [doc for score, doc in scored_docs[:2]]
        
        print(f"재정렬된 최고 점수: {scored_docs[0][0]}")

        # 최종 답변 생성
        context = "\n\n".join(doc.page_content for doc in top_docs)
        return self.final_chain.invoke({"context": context, "question": question})
```

```python
re_rank_chain = MapReRankChain(llm=llm, retriever=retriever)
```

```python
re_rank_chain.invoke("제비가 물어온 것은?")
```
