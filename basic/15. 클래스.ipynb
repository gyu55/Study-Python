```python
# 완전 객체 지향형 언어
# Market
# 판매
class Market:
    def __init__(self, market_name, product_name, product_price):
        self.market_name = market_name
        self.product_name = product_name
        self.product_price = product_price

    def sell(self, customer):
        customer.money -= self.product_price * (100 - customer.discount) / 100

# Customer
# 할인율
class Customer:
    def __init__(self, name, money, discount):
        self.name = name
        self.money = money
        self.discount = discount

        
```

```python
cu = Market("CU", "츄파츕스", 450)
customer = Customer("노규호", 4500, 10)

cu.sell(customer)
print(customer.money)
```

```python
# ATM, User 객체

# deposit 객체

# Atm.deposit(User, 10000)
# 유저의 돈은 10000 차감
# 은행의 돈은 10000 입금

# 은행이 가진 돈과 유저의 돈 출력
class Atm:
    def __init__(self, atm_money):
        self.atm_money = atm_money 
        
    def deposit(self, user, user_money):
        self.atm_money += user_money
        user.user_money -= user_money
        print(f"{user_money}원이 입금되었습니다.")

    def print_atm(self):
        print(f"은행이 가진 돈: {self.atm_money}")

class User:
    def __init__(self, user_money):
        self.user_money = user_money

    def print_user(self):
        print(f"유저가 가진 돈: {self.user_money}")

atm = Atm(1000000)
user = User(500000)


atm.print_atm()
user.print_user()
atm.deposit(user, 250000)
atm.print_atm()
user.print_user()


```

```python
# 클래스 메서드
class Korean:
    # 클래스 변수
    country = "한국"

    # 클래스 메서드
    # 인스턴스 X, 클래스변수 O
    @classmethod # 데코레이터
    def trip(cls, country):
        if cls.country == country:
            print("국내여행")
        else:
            print("해외여행")

    # 정적 메서드
    @staticmethod
    def slogan():
        print("Imagine your korea")
        

```

```python
hong = Korean()

hong.trip("한국")
hong.slogan()

Korean.trip("미국")
Korean.slogan()
```

# 상속(inheritance)

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age
    
```

```python
hong = Person("홍길동")
jang = Student("장보고", 20)

# 부모 -> 타입도 같이 전달!
# 모든 자식은 부모 타입이다.
# 부모는 자식 타입이 아니다.
print(isinstance(hong,Person))
print(isinstance(hong,Student))
print(isinstance(jang,Person))
print(isinstance(jang,Student))
```

```python
class Animal:
    def __init__(self, name, feed, species):
        self.name = name
        self.feed = feed
        self.species = species

    def feeding(self):
        print(f"{self.name}은 먹이를 먹습니다.")
        
class Dog(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)

    def act(self):
        print(f"{self.name}은 산책을 합니다.")

class Cat(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)

    def sleep(self):
        print(f"{self.name}은 잠을 자주 잡니다.")
```

```python
dog = Dog("누렁이", "잡식", "강아지")

dog.act()
dog.feeding()

cat = Cat("야옹이", "츄르", "고양이")

cat.sleep()
cat.feeding()
```

```python
# 사람(부모), 학생(자식)
# 사람 - 이름, 나이, 키, 몸무게
class Person:
    def __init__(self, name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

    # 사람은 인사하는 메서드
    @staticmethod
    def greeting():
        print(f"인사합니다.")

# 학생 - 학생, 핸드폰
class Student(Person):
    def __init__(self, name, age, height, weight, student_id, phone):
        super().__init__(name, age, height, weight)

    # 학생은 공부하는 메서드
    def study(self):
        print(f"{self.name}이(가) 공부합니다.")

hong = Person("홍길동", "15", "175", "75")
hong.greeting()

student_a = Student("장보고", "15", 175, 75, "20261234", "01012341234")
student_a.greeting()
student_a.study()
```

# 다형성

```python
class Car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price
        
    def start_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 켭니다.")

    def stop_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 끕니다.")
    
class SuperCar(Car):
    def start_engine(self):
        print(f"{self.brand}는 음성으로 시동을 켭니다.")
        
    def stop_engine(self):
        print(f"{self.brand}는 음성으로 시동을 끕니다.")
        
```

```python
matiz = Car("마티즈", 10000000)
matiz.start_engine()
matiz.stop_engine()
```

```python
# 휴대폰 필드
class Phone:
# 브랜드(brand), 가격(price)
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price
# 2G 전화하기
class Phone_2G(Phone):
    def __init__(self, brand, price):
        super().__init__(brand, price)

    def call(self):
        print(f"2G 전화하기")

# 3G 전화하기, 인터넷
class Phone_3G(Phone_2G):
    def __init__(self, brand, price):
        super().__init__(brand, price)

    def call(self):
        print(f"3G 전화하기")

    def internet(self):
        print(f"3G Internet")

# 4G 전화하기, 인터넷, 영상통화
class Phone_4G(Phone_3G):
    def __init__(self, brand, price):
        super().__init__(brand, price)

    def call(self):
        print(f"4G 전화하기")

    def internet(self):
        print(f"4G Internet")

    def facetime(self):
        print(f"4G 영상통화")
```

```python
phone_2g = Phone_2G("LG", 100_000)
phone_2g.call()

phone_3g = Phone_3G("Samsung", 500_000)
phone_3g.call()
phone_3g.internet()

phone_4g = Phone_4G("IPhone", 1_000_000)
phone_4g.call()
phone_4g.internet()
phone_4g.facetime()
```
