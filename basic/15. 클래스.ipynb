```python
# 이름(name), 나이(age), 운동종목(sports)
# 플레이어들을 소개하는 메서드

# 운동선수 3명의 객체화 하고
# 3명을 각각 소개하는 메서드를 사용해서 정보들을 출력하세요.
# ex) OOO선수의 나이 OO살 운동종목 OOOO
class Player:

    def __init__(self, name, age, sports):
        self.name = name
        self.age = age
        self.sports = sports

    def print_info(self):
        print(f"{self.name}선수의 나이 {self.age}살 운동종목 {self.sports}")

p1 = Player("이순신", 30, "수영")
p2 = Player("장영실", 40, "농구")
p3 = Player("춘식이", 15, "육상")
p1.print_info()
p2.print_info()
p3.print_info()
```

```python
# 완전 객체 지향형 언어
# Market
# 판매
class Market:
    def __init__(self, market_name, product_name, product_price):
        self.market_name = market_name
        self.product_name = product_name
        self.product_price = product_price

    def sell(self, customer):
        customer.money -= self.product_price * (100 - customer.discount) / 100

# Customer
# 할인율
class Customer:
    def __init__(self, name, money, discount):
        self.name = name
        self.money = money
        self.discount = discount

        
```

```python
cu = Market("CU", "츄파츕스", 450)
customer = Customer("노규호", 4500, 10)

cu.sell(customer)
print(customer.money)
```

```python
# ATM, User 객체

# deposit 객체

# Atm.deposit(User, 10000)
# 유저의 돈은 10000 차감
# 은행의 돈은 10000 입금

# 은행이 가진 돈과 유저의 돈 출력
class Atm:
    def __init__(self, atm_money):
        self.atm_money = atm_money 
        
    def deposit(self, user, user_money):
        self.atm_money += user_money
        user.user_money -= user_money
        print(f"{user_money}원이 입금되었습니다.")

    def print_atm(self):
        print(f"은행이 가진 돈: {self.atm_money}")

class User:
    def __init__(self, user_money):
        self.user_money = user_money

    def print_user(self):
        print(f"유저가 가진 돈: {self.user_money}")

atm = Atm(1000000)
user = User(500000)


atm.print_atm()
user.print_user()
atm.deposit(user, 250000)
atm.print_atm()
user.print_user()


```
