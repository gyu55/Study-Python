## 1. 리스트 컬렉션 메서드

```python
li1 = [ i + 1 for i in range(10) ]
```

```python
# 맨 뒤에 추가
li1.append(10)
```

```python
li1
```

```python
for i in range(11, 21):
    li1.append(i)
```

```python
li1
```

```python
# list의 모든 합 구하기
sum = 0
for i in li1:
    sum += i


print(sum)
```

```python
# .remove
li1.remove(3)
```

```python
li1.pop(1)
```

```python
len("abcd") # str
len([10, 20]) # list
len({10, 20}) # set
len({"a": 10, "b": 20}) # dict
```
