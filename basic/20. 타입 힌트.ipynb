```python
class User:
    name: str
    age: int
    address: str
    phone: str

    def __init__(self, name, age, address, phone):
        self.name = name
        self.age = age
        self.address = address
        self.phone = phone

    def print_hobby(self, hobby: str) -> None:
        print(f"{self.name}의 취미는 {hobby}입니다.")
```

```python
jang = User("장보고", 20, "서울시 종로구", "010-1234-4321")
```

```python
jang.address
```

```python
from typing import TypedDict
```

```python
class User(TypedDict):
    name: str
    address: str
    hobby: str
```

```python
user1_data: User = {
    "name": "홍길동",
    "address": "경기도 남양주시",
    "hobby": "코딩"
}
```

```python
User(**user1_data)
```

## Pydantic
- 파이썬에서 데이터를 검증하기 위한 라이브러리
- 타입 힌트와 동일하나, 

```python
from pydantic import BaseModel
```

```python
class User(BaseModel):
    name: str
    address: str
    hobby: str
```

```python
# 키가 더 추가되면 틀 이외 값은 무시
# 타입이 다를 때 오류를 검증함 (ValidationError)
user2_data: User = {
    "name": 10,
    "address": "경기도 구리시",
    "hobby": "코딩"
}
```

```python
User(**user2_data)
```
