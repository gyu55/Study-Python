```python
# 두 수를 더하고 2배 곱한 결과를 반환
# 콜백함수가 출력(print)일 경우 return 하면 안 됨
def add(num1, num2, callback):
    callback(num1 + num2)

def get_square(num, callback):
    callback(num * 2)
```

```python
# lambda
add(10, 20, lambda result: get_square(result, print))
```

```python
# 성과 이름을 전달하면 풀 네임으로 붙이고 "님"을 붙여서 반환

def get_full_name(last_name, first_name, callback):
    callback(last_name + first_name)
def add_nim(full_name, callback):
    callback(full_name + "님")

get_full_name("홍", "길동", lambda full_name: add_nim(full_name, print))
```

```python
# 두 수를 전달하면 두 수를 더하고, 2배를 만들고, 10을 더해서 리턴

def add(num1, num2, callback):
    return callback(num1 + num2)

def mulBy2(num, callback):
    return callback(num * 2)

def add_10(num):
    return num + 10

add(10, 20, lambda total: mulBy2(total, add_10))
```

## map, filter

```python
li1 = [(i + 1) * 10 for i in range(10)]
```

```python
li1
```

```python
def mul_2(num):
    return num * 2

# map(함수, 객체)
list(map(mul_2, li1))
# 람다함수는 값을 리턴
list(map(lambda i : i * 2, li1))
```

```python
# 65 ~ 90 ( A ~ Z )
li2 = [ i for i in range(65, 91)]
print(li2)
```

```python
li3 = list(map(lambda c: chr(c), li2))
print(li3)
```

```python
# li3 소문자로 변경 ( 97 ~ 121 )

def ord_num(num):
    return num + 32
li4 = list(map(lambda c: chr(c), list(map(ord_num, list(map(lambda c: ord(c), li3))))))
print(li4)


li4 = list(map(lambda c: chr(ord(c) + 32), li3))
print(li4)


li4 = list(map(lambda c : c.lower(), li3))
print(li4)
```

```python
# li2로 aBcDeFg...로 변경 후 li5를 생성
li5 = list(map(lambda c: chr(c + 32) if c % 2 == 1 else chr(c), li2))
result = "".join(li5)
print(result)

# li5 = list(map(lambda c : (chr(c) if c%2 == 0 else chr(c).lower()), li2))
# print(li5)
```

```python
li6 = [i + 1 for i in range(100)]
print(li6)
```

```python
# 3의 배수
li7 = list(filter(lambda n: n % 3 == 0, li6))
print(li7)
```

```python
# li6에 3의 배수들만 걸러내고 2배로 값을 바꾸기
li8 = list(map(lambda n: n * 2, filter(lambda n: n % 3 == 0, li6)))
print(li8)
```

```python
# li4에서 a ~ g까지만 filter로 걸러내고 li9 만들기

li9 = list(filter(lambda n: 'a' <= n <= 'g', li4))
print(li9)


li9 = list(map(lambda tu: tu[1], filter(lambda tu: tu[0] < 7, enumerate(li4))))
print(li9)
```
