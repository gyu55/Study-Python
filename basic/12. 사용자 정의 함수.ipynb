```python
# 구매한 상품의 총 가격과 구매자의 이름을 리턴해주는 함수

def order_info(*args, name, address, phone):
    total = sum(args)
    print(f"{name}님의 주소: {address}")
    print(f"{name}님의 전화번호: {phone}")
    print(f"{name}님의 총 구매 가격: {total}")
```

```python
# 두 값의 사칙연산 중 계산하고 싶은 값만 리턴해주기
# calc가 있을 때에만 계산해서 리턴, 없을 때에는 (10, 20)을 튜플 형태로 리턴

# 1. 두 개의 변수는 따로 변수에 받기
# 2. 부호는 따로 kwards 문법으로 받기

# ex) 10, 20, "+"

def calculator(num1: int, num2:int, **kwards) -> float:

    calc = kwards.get("calc")
    if calc is None:
        return (num1, num2)
    
    return eval(f"{num1} {calc} {num2}")

    
```

```python
print(calculator(10,20))
print(calculator(10,20, calc="+"))
print(calculator(10,20, calc="-"))
print(calculator(10,20, calc="*"))
print(calculator(10,20, calc="/"))
print(calculator(10,20, calc="**"))
```

```python
# 글자의 개수를 찾아주는 함수
# ex) 
# "apple", "a" -> 1
# "apple", "p" -> 2

def find_word_count(words: str, find_word: str) -> int:
    return words.count(find_word)
words = input("단어를 입력하세요.\n")
find_word = input("찾을 글자를 입력하세요\n")
print(f"찾는 글자: {find_word}\n 개수:{find_word_count(words, find_word)}")
```

```python

def char_count(word: str) -> dict:
    di_char = {}
    for c in word:
        if c in di_char:
            di_char[c] += 1
        else:
            di_char[c] = 1
    return di_char

char_count("apple")
```

```python
def char_count_update(word: str) -> dict:
    result = {}

    for c in word:
        result.update({c : word.count(c)})
    return result
char_count_update("apple")
```
