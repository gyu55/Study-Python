```python
class User:
    name: str
    age: int
    phone: str

    
    def __init__(self, name, age, phone):
        self.name = name
        self.age = age
        self.phone = phone
        
    # toString 재정의
    def __str__(self):
        return f"User(name: {self.name}, age: {self.age}, phone: {self.phone})"
```

```python
print(User("홍길동", 20, "010-1234-1234"))
```

## 연산자

```python
# 회사원
class Employee(BaseModel):
    id: int
    name: str
    income: int
```

```python
employee_data1 = {
    "id": 1,
    "name": "홍길동",
    "income": 10000
}
employee_data2 = {
    "id": 2,
    "name": "이순신",
    "income": 30000
}
employee1 = Employee(**employee_data1)
employee2 = Employee(**employee_data2)
```

```python
print(employee1)
print(employee2)
```

```python
class Student(BaseModel):
    id: int
    name: str
    income: int
```

```python
student_data1 = {
    "id": 1,
    "name": "김철수",
    "income": 10000,
}
student_data2 = {
    "id": 2,
    "name": "김영희",
    "income": 20000,
}

student1 = Student(**student_data1)
student2 = Student(**student_data2)
```

```python
print(employee1)
print(employee2)
print(student1)
print(student2)
```

```python
print(Employee.__eq__(employee1, employee1))
print(Employee.__eq__(employee1, employee2))
print(Student.__eq__(student1, student1))
print(Student.__eq__(student1, student2))
```

```python
student1 = Student(**student_data1)
student3 = Student(**student_data1)

print(student1)
print(student3)
```

```python
print(Student.__eq__(student1, student3))
```

```python
# 문제가 생길 수 있는 코드
# equals 재정의 필요
class Student:
    def __init__(self, id, name):
        self.id = id
        self.name = name

    def __eq__(self, other):
        # 해시 주소 비교
        if self.__repr__() == other.__repr__():
            return True

        # 타입비교
        if isinstance(other, Student):
            # id 끼리 값 비교
            return self.id == other.id
        return False
```

```python
student4 = Student(1, "홍길동")
student5 = Student(1, "홍길동")

print(Student.__eq__(student4, student5))
print(student4 == student5)
```
