# 컬렉션(list, tuple, set, dict)
1. list []
    - 추가, 수정, 삭제 언제나 가능
    - [] 대괄호로 묶어주고 ,(콤마)로 값을 구분한다

2. tuple ()
   - 한 번 생성되고나면 변경이 불가능
   - () 소괄호로 묶어주고 ,(콤마)로 값을 구분한다

3. set {}
   - 순서가 없는 자료 구조(비시퀀스)
   - 중복된 값을 저장이 불가능
   - {} 중괄호로 묶어주고 ,(콤마)로 값을 구분한다

4. dict {key:value}
   - 키 + 값이 한 쌍으로 관리하며 ,(콤마)로 값을 구분한다
   - key는 중복 값을 허용하지 않으며, value는 중복값을 허용한다
   - key를 통해 value를 가져온다

# 딕셔너리 

```python
# {}: dict 
set1 = {}
set2 = set()
di1 = {}
```

```python
print(type(set1))
print(type(set2))
print(type(di1))
```

```python
di2 = {
    "a" : "apple",
    "b" : "banana",
    "c" : "cherry",
    "a" : "alpha"
}
```

```python
# key가 있으면 수정, key가 없으면 추가 -> key의 중복값을 허용하지 않음
di2
```

```python
# dict에서 key를 통해 value를 가져오는 방법
a = 'a'
di2[a]
di2['a']
di2.get('b')
```

```python
# di2["d"] -> 없는 값을 가져오는 오류

# 권장 -> 값이 없을 경우 Optional로 가져올 수 있음
di2.get("d", "없음(초기값)")
```

```python
print(di2.keys())
print(di2.values()) # -> list는 아님
print(di2.items())

print(type(di2.keys()))
```
